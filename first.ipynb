```python
import csv
from cffi.backend_ctypes import xrange
```

Функции сортировки

```python
def partitionx(array, begin, end):
    pivot = begin
    for i in xrange(begin + 1, end + 1):
        if array[i][0] <= array[begin][0]:
            pivot += 1
            array[i], array[pivot] = array[pivot], array[i]
    array[pivot], array[begin] = array[begin], array[pivot]
    return pivot
```

```python
def quick_sortx(array, begin=0, end=None):
    if end is None:
        end = len(array) - 1

    def _quicksortx(array, begin, end):
        if begin >= end:
            return
        pivot = partitionx(array, begin, end)
        _quicksortx(array, begin, pivot - 1)
        _quicksortx(array, pivot + 1, end)
    return _quicksortx(array, begin, end)
```

Функция для нахождения медианы

```python
def mediana(array):
    n = len(array)
    index = n // 2
    if n % 2:
        return sorted(array)[index]
    else:
        return sum(sorted(array)[index - 1:index + 1]) / 2
```

Считываем данные и записываем их в словарь shops

```python
with open ("milk_promo_sales.csv", 'r') as milk:
    new_ref = csv.reader(milk, delimiter=",")
    shops = {}  # словарь, в котором ключами являтся номера магазинов, а значениями двумерные списки,
    # в которых первый элемент - номер недели наблюдения, а второй - объем продаж в эту неделю
    ids=[] # ids - это номера всех магазинов
    count = 0
    for string in new_ref:
        buffer=[] # список со всеми неделями наблюдения и соответствующими продажами
        if count != 0:
            if int(string[0]) not in shops: # если строчка с номером данного магазина не попадалась ранее, 
                # вносим ее в словарь
                ids.append(int(string[0]))
                shops[int(string[0])]=[[int(string[1]),float(string[2])]]
            else:  # если строчка с номером данного магазина уже попадалась, 
                # добавляем значение для соответствующего ключа в словаре 
                buffer=shops[int(string[0])]
                buffer.append([int(string[1]),float(string[2])])
                shops[int(string[0])]=buffer
        count+=1
```

```python
ids = sorted(ids) 

kolvo = 0 # kolvo - искомое количество всех промопериодов

kolvosloc = [] # kolvosloc - список с количеством промопериодов на каждый магазин 

dlina = 0 # dlina - длина рассматриваемого промопериода

dliny = [] # dliny - список с длинами всех промопериодов

number_promo=1 # number_promo - номер промопериода

sumloc = 0 # sumloc - сумма продаж в рассматриваемом промопериоде

sumsloc = [] # sumsloc - список с суммами продаж за все промопериоды
```

Находим соответствующие каждому магазину промопериоды, продажи и тд  и вычислим искомые величины

```python
for nomer in ids:
    quick_sortx(shops[nomer]) # сортируем все пропопериоды по неделям для данного магазина
    magazin = shops[nomer]
    for znach in range(len(magazin)):
        dlina += 1
        sumloc += magazin[znach-1][1]
        if len(magazin) == znach + 1 or magazin[znach + 1][0] - magazin[znach][0] > 1:
            dliny.append(dlina)
            sumsloc.append([nomer, number_promo, sumloc])
            dlina = 0
            kolvo += 1
            sumloc = 0
            number_promo += 1
    kolvosloc.append(kolvo-sum(kolvosloc))
med_kolvo = mediana(kolvosloc)
med_dlin = mediana(dliny)
```
